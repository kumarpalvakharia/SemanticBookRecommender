```python
books['missing_description'] = np.where(books['description'].isna(),1,0)
books['age_of_books'] = 2025-books['published_year']
```

```python
columns_of_interest = ['num_pages','age_of_books','missing_description','average_rating']
```

```python
corr_matrix = books[columns_of_interest].corr(method='spearman')
```

```python
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar_kws={"label": "Spearman Correlation Coefficient"})
plt.show()
```

```python
book_missing = books[~(books['description'].isna())  & 
      ~(books['published_year'].isna()) &
      ~(books['num_pages'].isna()) &
      ~(books['average_rating'].isna()) ]
```

```python
book_missing
```

```python
book_missing['categories'].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing['words_in_description'] = book_missing['description'].apply(lambda x: len(str(x).split()))
```

```python
book_missing_25_words = book_missing[book_missing['words_in_description']>=25].copy()
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle","missing_description","age_of_books","words_in_description"],axis=1)
    .to_csv("book_cleaned.csv", index=False)
)
```
